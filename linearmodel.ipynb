```python
#IMPORTING DEPENDENCIES
import pandas as pd
import numpy as np
import tensorflow as tf
import csv
from tensorflow import keras
from keras.utils import to_categorical
from keras.models import Model
from keras.layers import Dense, Input, Dropout, LSTM, Activation, Bidirectional,BatchNormalization
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence
from keras.initializers import glorot_uniform
import pickle 
import nltk
```

```python
###############################################################################################################################
#READING CSV INTO X_TRAIN AND Y_TRAIN
df=pd.read_csv('train1.csv')
print(df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(min_df=2,max_df=5,ngram_range=(1,2))
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
```

```python
sentences=df['Text'].tolist()
new_sentences=[]
for sentence in sentences:
    new_sentence=[]
    for word in sentence.split():
        if word in stop_words:
            continue
        else:
            new_sentence.append(word)
    new_sentence=' '.join(new_sentence)
    new_sentences.append(new_sentence)
sentences=new_sentences
features=tfidf.fit_transform(sentences)
dffeatures=pd.DataFrame(features.todense(),columns=tfidf.get_feature_names())
x_train=np.array(features.todense())
```

```python
dffeatures
```

```python
y_train=np.array(df['Label'].tolist()).reshape(-1,1)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
score = lr.score(x_train, y_train)
print(score)
```

```python
from sklearn import svm
svc = svm.SVC(kernel='linear', C=1).fit(x_train, y_train)
```

```python
svc.score(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(random_state=2,max_depth=50)
```

```python
tree.fit(x_train,y_train)
```

```python
tree.score(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```
